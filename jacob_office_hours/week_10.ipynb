### PCA

```python
from sklearn.datasets import fetch_lfw_people
```

```python
people = fetch_lfw_people(min_faces_per_person=30)
```

```python
people.images[0]
```

```python
plt.imshow(people.images[0])
plt.title(people.target_names[0]);
```

```python
X = people.data/255
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(20)
pca2 = PCA(0.8)
```

```python
Xt = pca.fit_transform(X)
Xtb = pca2.fit_transform(X)
```

```python
pca.explained_variance_
```

```python
Xtb.shape
```

```python
plt.scatter(Xt[:, 0], Xt[:, 1], alpha = 0.2)
```

```python
Xt
```

```python
X.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
y = people.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xtb, y)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
image_shape = people.images[0].shape
```

```python
plt.imshow(pca.mean_.reshape(people.images[0].shape), cmap = plt.cm.bone)
```

```python
fig, axes = plt.subplots(3, 5, figsize=(15, 12),
                         subplot_kw={'xticks': (), 'yticks': ()})
for i, (component, ax) in enumerate(zip(pca2.components_, axes.ravel())):
    ax.imshow(component.reshape(image_shape),
              cmap='viridis')
    ax.set_title("{}. component".format((i + 1)))
```
